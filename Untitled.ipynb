```python
data = pd.read_csv('lastfm.csv')
data.head()
```

```python
data_germany = data.drop('user', axis = 1)
```

```python
data_ibs = pd.DataFrame(index = data_germany.columns, columns= data_germany.columns)
```

```python
for i in range(len(data_ibs)):
    for j in range(len(data_ibs)):
        data_ibs.ix[i, j] = 1 - cosine(data_germany.ix[:,i], data_germany.ix[:,j])
```

```python
data_ibs
```

```python
data_neigbhors = pd.DataFrame(index = data_ibs.columns, columns= range(1, 11))
```

```python
for i in range(len(data_neigbhors)):
    data_neigbhors.ix[i, :] = data_ibs.ix[:, i].sort_values(ascending = False)[1:11].index
```

```python
documents = (
"The sky is blue",
"The sun is bright",
"The sun in the sky is bright",
"We can see the shining sun, the bright sun"
)
```

```python
tfidf_matrix
```

```python
from sklearn.metrics.pairwise import cosine_similarity
cosine_similarity(tfidf_matrix[0:1], tfidf_matrix)
```
